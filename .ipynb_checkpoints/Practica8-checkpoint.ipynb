# Práctica 8: Obtención de datos desde archivos

En el ambiente de programación MATLAB es posible generar matrices y tablas de datos a partir de archivos en los formatos **.txt** y **.CSV**, este último formato se refiere a datos provenientes de un archivo de software de adquisición de datos o incluso de excell.


```matlab
clear 
close all
datos = readtable('https://raw.githubusercontent.com/LuisGerardo2204/Archivos_de_datos/main/ADC.txt');
```

```matlab
V = table2array(datos);
```

```matlab
t=V(:,1);
v1=V(:,2);
v2=V(:,3);
v3=V(:,4);
plot(t,v1);
hold on
plot(t,v2);
plot(t,v3);
```

```matlab
clear
close all
nombre_archivo = 'Datos_web.txt';
websave(nombre_archivo,'https://raw.githubusercontent.com/LuisGerardo2204/Archivos_de_datos/main/ADC.txt');

V2=load('Datos_web.txt');
t=V2(:,1);
v1=V2(:,2);
v2=V2(:,3);
v3=V2(:,4);
plot(t,v1);
hold on
plot(t,v2);
plot(t,v3);
```

```matlab
base=[2:2:15]*2.54;
altura=[2:2:18]*2.54;
[altura_nueva,base_nueva]=meshgrid(altura,base);
areas=0.5*altura_nueva.*base_nueva;

% Tabla de conversión de centímetros a pulgadas, pies y yardas
tabla=[base;areas']; % Se agrupan los datos convertidos 
%en una matriz que será la tabla de datos

% Finalmente, se imprime en la pantalla la tabla de equivalencias
disp("  Tabla de areas ")
fprintf(" base   %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f   %6.3f   %6.3f  %6.3f",altura)
fprintf("\n")
fprintf("%6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f  %6.3f      \n",tabla)
```
